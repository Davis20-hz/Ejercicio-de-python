```python
def runningSum(nums):
    results = [0] * len(nums)
    results[0] = nums[0]

    for i in range(1, len(nums)):
        results[i] = nums[i] + results[i - 1]

    return results


# Entrada del usuario:
entrada = input("Ingresa la lista, por ejemplo [1,2,3,4]: ")

# Convierte la entrada tipo texto a una lista real
nums = eval(entrada)

resultado = runningSum(nums)
print(resultado)
```
